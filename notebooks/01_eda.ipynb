```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

pd.set_option("display.max_columns", 500)
pd.set_option("display.max_rows", 200)

sns.set_style("whitegrid")
```

```python
DATA_DIR = Path("../data/raw")

train_trans_path = DATA_DIR / "train_transaction.csv"
train_id_path = DATA_DIR / "train_identity.csv"
test_trans_path = DATA_DIR / "test_transaction.csv"
test_id_path = DATA_DIR / "test_identity.csv"
```

```python
train_transaction = pd.read_csv(train_trans_path)
train_identity = pd.read_csv(train_id_path)
test_transaction = pd.read_csv(test_trans_path)
test_identity = pd.read_csv(test_id_path)
```

```python
print("Train Transaction Shape:", train_transaction.shape)
print("Train Identity Shape:", train_identity.shape)
print("Test Transaction Shape:", test_transaction.shape)
print("Test Identity Shape:", test_identity.shape)
```

```python
train_df = train_transaction.merge(train_identity, on="TransactionID", how="left")
test_df = test_transaction.merge(test_identity, on="TransactionID", how="left")

print("Merged Train Shape:", train_df.shape)
print("Merged Test Shape:", test_df.shape)
```

```python
train_df.head()
```

```python
fraud_ratio = train_df["isFraud"].mean()

print(f"Fraud Ratio: {fraud_ratio:.6f}")
print(f"Fraud Percentage: {fraud_ratio * 100:.4f}%")
```

```python
print("Total columns:", len(train_df.columns))
train_df.columns.tolist()[:100]
```

```python
missing_percent = train_df.isnull().mean().sort_values(ascending=False) * 100
missing_df = pd.DataFrame({
    "column": missing_percent.index,
    "missing_percent": missing_percent.values
})

missing_df.head(30)
```

```python
top_missing = missing_df.head(20)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_missing, x="missing_percent", y="column")
plt.title("Top 20 Columns by Missing Percentage")
plt.xlabel("Missing Percentage")
plt.ylabel("Column")
plt.show()
```

```python
train_df["TransactionAmt"].describe()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(train_df["TransactionAmt"], bins=100, kde=False)
plt.title("Transaction Amount Distribution")
plt.xlim(0, 5000)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(np.log1p(train_df["TransactionAmt"]), bins=100, kde=False)
plt.title("Log-Transformed Transaction Amount Distribution")
plt.show()
```

```python
important_cols = ["TransactionDT", "TransactionAmt", "ProductCD", "card1", "card2", "card3", "card4", "card5", "card6"]

train_df[important_cols].head()
```

```python
fraud_by_product = train_df.groupby("ProductCD")["isFraud"].mean().sort_values(ascending=False)
fraud_by_product
```

```python
plt.figure(figsize=(8, 4))
fraud_by_product.plot(kind="bar")
plt.title("Fraud Rate by ProductCD")
plt.ylabel("Fraud Rate")
plt.show()
```

```python
cat_cols = ["ProductCD", "card4", "card6", "P_emaildomain", "R_emaildomain", "DeviceType"]

for col in cat_cols:
    if col in train_df.columns:
        print(f"{col}: {train_df[col].nunique(dropna=False)} unique values")
```

```python
train_df["TransactionDT"].describe()
```

```python
train_df["TransactionDT_days"] = train_df["TransactionDT"] // (24 * 60 * 60)
train_df["TransactionDT_hours"] = train_df["TransactionDT"] // (60 * 60)

train_df[["TransactionDT", "TransactionDT_days", "TransactionDT_hours"]].head()
```

```python
fraud_by_day = train_df.groupby("TransactionDT_days")["isFraud"].mean()

plt.figure(figsize=(12, 5))
fraud_by_day.plot()
plt.title("Fraud Rate Over Day Buckets")
plt.xlabel("Day Bucket")
plt.ylabel("Fraud Rate")
plt.show()
```
